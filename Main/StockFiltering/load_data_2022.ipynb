```python
# Tải dữ liệu Q3-2018 của các công ty niêm yết trên 3 sàn chứng khoán
# thêm cột thông tin ngành và lưu vào file raw data 'data_Q3-2018-mergedIndustry.csv'
```

```python
# import thư viện cần thiết
import vnstock as vnst
import pandas as pd
import numpy as np
```

```python
# Tạo danh sách mã ticker của tất cả các công ty niêm yết
stock_list = vnst.listing_companies()
stock_list1 = pd.DataFrame(stock_list)
stocks = tuple(stock_list1['ticker'])
stocksArray = np.asarray(stocks)
stocksArray
```

```python
# Tạo dataframe để lưu thông tin của tất cả các công ty khi gọi hàm financial_ratio()
df = pd.DataFrame(columns=['ticker', 'quarter', 'year', 'priceToEarning', 'priceToBook', 'roe', 'roa',
                           'earningPerShare', 'bookValuePerShare', 'interestMargin', 'nonInterestOnToi',
                           'badDebtPercentage', 'provisionOnBadDebt', 'costOfFinancing', 'equityOnTotalAsset',
                           'equityOnLoan', 'costToIncome', 'equityOnLiability', 'epsChange', 'assetOnEquity',
                           'preProvisionOnToi', 'postTaxOnToi', 'loanOnEarnAsset', 'loanOnAsset', 'loanOnDeposit',
                           'depositOnEarnAsset', 'badDebtOnAsset', 'liquidityOnLiability', 'payableOnEquity',
                           'cancelDebt', 'bookValuePerShareChange', 'creditGrowth'])
```

```python
# Vòng lặp để lấy dữ liệu từ phương thức financial_ratio cho từng công ty và thêm vào DataFrame
for ticker in stocksArray:
    try:
        print("Loading company info ticker ", ticker)
        ptcoban = vnst.financial_ratio(ticker, 'quarterly', True)
        if 'Q3-2022' in ptcoban.columns:
            df.loc[len(df)] = ptcoban['Q3-2022']
    except KeyError:
        print("Skip load ", ticker)
```

```python
len(df)
```

```python
# Lấy thông tin tên ngành và merge với dataframe ở trên
List = vnst.listing_companies()
df_mergeIndus = pd.merge(df, List[['ticker', 'industry', 'organName']], on='ticker', how='left')
df_mergeIndus
```

```python
# Lưu lại dataframe vào file .csv
df_mergeIndus.to_csv('data_Q3-2022-mergedIndustry.csv')
```
